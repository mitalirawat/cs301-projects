```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import requests
import json
```

```python
# question1 - write a function to get the data from a file "filename" at url "url" 
#- check the contents of this file- u will see that they are names of some json file
#send a get request on each of these json files and collect the result as a list of dictionaries
#load the dict into a dataframe
# process countries data
def get_data_from_file(url, filename):
    r = requests.get(url+filename)
    filelist = r.text
    #print(filelist)
    data_list = []
    url_list = filelist.split('\n')
    #print(url_list)
    for file in url_list:
        if file =="":
            continue
        resp = requests.get(url+file)
        #either add to dict or to dataframe
        d = json.loads(resp.text)
        data_list.append(d)
    cntry_data = pd.DataFrame.from_dict(data_list)
    return cntry_data
        
```

```python
# Q2- write a function to get data from the given two urls and file names
country_data = get_data_from_file("http://localhost:8082/Capitals_with_Data/", "file_list.txt")
country_capitals = get_data_from_file("http://localhost:8082/Countries_with_Capitals/", "file_list.txt")
```

```python
# Q-3 Write a function to clean the 2 dataframes
# Rename column - "Captial Longitude" to "Capital Longitude"
# delete the region column
country_capitals.rename(columns={'Captial Longitude':'Capital Longitude'}, inplace=True)
country_data = country_data.drop(columns=['Region','Phones (per 1000)' ])
```

```python
#Also many columns contains commas (,) instead of (.) decimal points- fix so that the data is a a valid decimal number
country_data['Literacy rate'] = country_data['Literacy rate'].str.replace(',', '.')
#lit = [float((k).replace(',','.')) for k in list(country_data['Literacy rate'])]
country_data['Birthrate'] = country_data['Birthrate'].str.replace(',', '.')
country_data['Deathrate'] = country_data['Deathrate'].str.replace(',', '.')
country_data['Coastline (coast/area ratio)'] = country_data['Coastline (coast/area ratio)'].str.replace(',', '.')
country_data['Infant mortality (per 1000 births)'] = country_data['Infant mortality (per 1000 births)'].str.replace(',', '.')
country_data
```

```python
# Convert all numerical columns to float except population
country_data = country_data.astype({"Area (sq. mi.)": float,
                                    "Birthrate": float, "Coastline (coast/area ratio)": float, "Deathrate":float,
                                    "GDP ($ per capita)":float, "Infant mortality (per 1000 births)": float,
                                    "Literacy rate": float, "Population":int})
country_capitals = country_capitals.astype({"Capital Latitude":float,"Capital Longitude":float})
```

```python
#Query1: Find all Asian countries with population > threshold
def select_asian_countries(conn, region, population):
    cur = conn.cursor()
    cur.execute("SELECT Country, Population FROM country WHERE Continent = ? AND Population > ?", (region, population))
 
    rows = cur.fetchall()
 
    for row in rows:
        print(row)
```

```python
select_asian_countries(conn, 'Asia',200000000)
```

```python
#Query 2: List the continents in increasing order of infant mortality rates
def min_IMR_countries(conn, no_of_countries):
    cur = conn.cursor()
    cur.execute("SELECT Continent, AVG(`Infant mortality (per 1000 births)`) FROM country GROUP BY Continent ORDER BY AVG(`Infant mortality (per 1000 births)`) ASC LIMIT ?", (no_of_countries,))
 
    rows = cur.fetchall()
 
    for row in rows:
        print(row)
```

```python
min_IMR_countries(conn, 5)
```

```python
#Query 3: Find the largest country in each continent
def largestCountry(conn):
    cur = conn.cursor()
    cur.execute("SELECT Continent, Country, MAX(`Area (sq. mi.)`) FROM country GROUP BY Continent")
 
    rows = cur.fetchall()
 
    for row in rows:
        print(row)
```

```python
largestCountry(conn)
```

```python
#Query 4: Find the counties with growthrate> threshold and lying within given latitudes.
#Insert new column for growthrate as birthrate- deathrate
def min_birthrate_latitude(conn, a, b, c):# Top regions with the minimum IMR rates
    cur = conn.cursor()
    cur.execute("ALTER TABLE country ADD Growthrate integer")
    cur.execute("UPDATE country SET Growthrate=Birthrate-Deathrate")
    cur.execute("SELECT Country, Capital FROM country WHERE `Capital Latitude` > ? AND `Capital Latitude` < ? AND Growthrate > ?",(a,b, c))
 
    rows = cur.fetchall()
 
    for row in rows:
        print(row)
```

```python
min_birthrate_latitude(conn, -5, 1, 30)
```
